W jakim miesiącu oddano najwięcej głosów?
Format YYYY-MM. np. 2022-05

```python
votes = pd.read_xml('../03-pandas/data/travel.stackexchange.com/Votes.xml', parser='etree')
```

```python
votes.head()
```

```python
votes['CreationDate'] = pd.to_datetime(votes['CreationDate'])
votes['Months'] = votes['CreationDate'].dt.to_period('M')
```

```python
votes.head()
```

```python
months_df = votes['Months'].value_counts()
months_df
```

Na przełomie jakich dwóch miesiecy nastąpił największy spadek ilości oddanych głosów?
Podaj drugi z tych miesięcy. Format YYYY-MM. np. 2022-05

```python
months_df = votes['Months'].value_counts().rename_axis('dates').reset_index(name='counts').sort_values('dates')
months_df
```

```python
months_df[1:]
```

```python
losses = [0,]
prev_count = months_df.iloc[0, :].loc["counts"]

for index, row in months_df[1:].iterrows():
    curr_count = row["counts"]
    losses.append(curr_count - prev_count)
    prev_count = curr_count

months_df = months_df.assign(loss=losses)
    
```

```python
months_df
```

```python
months_df.sort_values('loss')
```

Jaki tag (patrzymy tylko na Posts) jest najpopularniejszy wśród użytkowników z Polski?
Odpowiedź podaj bez nawiasów <>. Za użytkownika z Polski uważamy kogoś kogo pole Location zawiera Poland lub Polska (moze zawierać też inne słowa, np. Olsztyn, Poland).

```python
polish_users = users[(users['Location'].str.contains('Poland|Polska', na=False))]
polish_users.head()
```

```python
users_post = posts.merge(
    polish_users,
    left_on = 'OwnerUserId',
    right_on="Id"
).loc[:, ["Id_y", "OwnerUserId", "Body", "Location", "Tags"]]


```

```python
users_post
```

```python
import re
CLEANR = re.compile('<.*?>') 

def cleanhtml(raw_html):
    cleantext = re.sub(CLEANR, '', raw_html)
    return cleantext
```

```python
all_tags = []
for tag in users_post["Tags"]:
    if tag is None:
        continue
    new_tag = str(tag).replace("<", "").replace(">", " ")
    tags = new_tag.split()

    all_tags.extend(tags)

all_tags

tags_df = pd.DataFrame(all_tags, columns=['tag'])
tags_df.value_counts()
```

Jakie słowo o długości powyżej 7 znaków pojawia się najczęściej w treści (Body) postów?

```python
posts.columns
```

```python
import re
CLEANR = re.compile('<.*?>') 

def cleanhtml(raw_html):
    cleantext = re.sub(CLEANR, '', raw_html)
    return cleantext
```

```python
posts["Body"]
```

```python
all_words = []
for post in posts["Body"]:
    body = cleanhtml(str(post))

    words = body.split()
    for word in words:
        if len(word) > 7:
            all_words.append(word)


words_df = pd.DataFrame(all_words, columns=['word'])
words_df.value_counts().head(5)
```
